```python
# Problem 1
# Factorial (1 point)
# The factorial of a number is the product of that number and all the numbers below it. For example, 5 factorial (denoted 5! in math) is 5 4 3 2 1, which is 120.

# Write a function called problem_1 that takes one positional argument number. The function should return the factorial of number as an integer.

# Example input/output:
# problem_1(5) => 120
# problem_1(10) => 3628800

print('Problem 1: This program will calculate for the factorial based on the given values of the problem.') # Program description

def problem_1(number): # Parameter
    factorial = 1 # If set to zero, the value of 'factorial' will always be zero.
    
    if number < 0: # We will only be considering positive values.
        print('This factorial calculator does not calculate for the factorial of negative numbers.')
    elif number == 0: # Mathematical definition.
        print(f"The factorial of {int(number)} is equivalent to 1.")
    else: # All possible exemptions have been mentioned.
        for value in range(1, (number)+1): # Begin at one and include the value of problem_1.
            # print(problem_1) Displays all the values we are dealing with (only for checking purposes).
            # problem_1 -= 1 Decrements the value of problem_1 by one (only for checking purposes).
            factorial = factorial * value # Formula of factorials.
        return factorial
        
    
print(problem_1(5)) # Argument 1
print(problem_1(10)) # Argument 2

# Reference: https://www.programiz.com/python-programming/examples/factorial 
```

```python
# Problem 2
# Guessing Game (2 points)
# Recall that we use input() to ask for user input.

# Write a function called problem_2 that takes one positional argument passphrase. The function should ask the user to enter the passphrase with input(). If the user enters the correct passphrase, the function should return the integer 1. If the user enters the wrong passphrase, the function should keep asking them for the passphrase. If the user gets the passphrase wrong 3 times, the function should return the integer 0.

# Example input/output:

# Function input:
# problem_2("chums")
# Simulated user input:
# joben
# chums
# Output:
# 1

# Function input:
# problem_2("annika")
# Simulated user input:
# joe
# earl
# niks
# Output:
# 0

print('''
Problem 2: This program will provide the user with three chances to guess the correct passphrase. 
If the user inputs the correct passphrase within three tries, the integer 1 will appear. 
Otherwise, the integer 0 will be displayed.''') # Program description

def problem_2(passphrase):
    guess_count = 0
    guess_limit = 3
    while guess_count < guess_limit:
        passphrase = str(input('Guess the passphrase: '))
        guess_count += 1
        if passphrase == 'chums':
            return int(True)
            break
    else:
        return int(False)

problem_2('chums')

# Reference: https://www.youtube.com/watch?v=_uQrJ0TkZlc
```
